```python
def train_linear_model(data):
    X_train, X_test, y_train, y_test = data
    model = LinearRegression()
    model.fit(X_train, y_train)
    return model.score(X_test, y_test)
```

```python
from sklearn.metrics import r2_score
from multiprocessing import Pool, cpu_count
import time

start_time = time.time()
score_normal = train_linear_model((X_train, X_test, y_train, y_test))
normal_duration = time.time() - start_time

print(f"Normal Training -> RÂ² Score: {score_normal:.4f}, Time: {normal_duration:.2f} seconds")
```

```python
def train_and_score(_):
    model = LinearRegression()
    model.fit(X_train, y_train)
    return r2_score(y_test, model.predict(X_test))

start_time = time.time()
with Pool(cpu_count()) as pool:
    scores_parallel = pool.map(train_and_score, range(cpu_count()))
score_parallel = sum(scores_parallel) / len(scores_parallel)
time_parallel = time.time() - start_time

print(f"Normal RÂ²: {score_normal:.4f} | Time: {normal_duration:.4f}s")
print(f"Parallel Avg RÂ²: {score_parallel:.4f} | Time: {time_parallel:.4f}s")
```

```python
from multiprocessing import Pool, cpu_count


def parallel_training():
    # Create multiple jobs of the same model training (to simulate CPU load)
    tasks = [(X_train, X_test, y_train, y_test)] * cpu_count()

    # Pool to distribute the jobs
    with Pool(cpu_count()) as pool:
        results = pool.map(train_linear_model, tasks)

    return np.mean(results)

start_time = time.time()
parallel_score = parallel_training()
parallel_duration = time.time() - start_time

print(f"Parallel Training -> Avg RÂ² Score: {parallel_score:.4f}, Time: {parallel_duration:.2f} seconds")
```

```python
import lightgbm as lgb
import xgboost as xgb
from multiprocessing import Pool, cpu_count
from joblib import Parallel, delayed



def train_lightgbm_model(X_train, y_train):
    model = lgb.LGBMRegressor(n_estimators=100,random_state=42, n_jobs=1) # Set n_jobs=1 for single core training in this function
    model.fit(X_train, y_train)
    return model

def train_xgboost_model(X_train, y_train):
    model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.05, max_depth=7, subsample=0.7, colsample_bytree=0.7, random_state=42, n_jobs=1) # Set n_jobs=1 for single core training
    model.fit(X_train, y_train)
    return model

# Parallel processing function adapted for different models
def parallel_processing_models(X_train, y_train, X_test, train_func, n_jobs=4):
    start_time = time.time()

    # Split data into chunks
    X_splits = np.array_split(X_train, n_jobs)
    y_splits = np.array_split(y_train, n_jobs)

    # Use joblib for parallel training
    models = Parallel(n_jobs=n_jobs)(
        delayed(train_func)(X, y) for X, y in zip(X_splits, y_splits)
    )

    # Ensemble predictions (mean for regression)
    preds = np.mean([model.predict(X_test) for model in models], axis=0)

    return {
        'time': time.time() - start_time,
        'predictions': preds
    }

# Normal sequential processing function adapted for different models
def normal_processing_models(X_train, y_train, X_test, train_func):
    start_time = time.time()
    model = train_func(X_train, y_train)
    preds = model.predict(X_test)
    return {
        'time': time.time() - start_time,
        'predictions': preds
    }

# Define the number of jobs for parallel processing
n_jobs = cpu_count() # Use the number of available CPU cores

# --- Compare LightGBM ---
print("--- Comparing LightGBM Processing ---")
# Normal LightGBM
normal_lgb_result = normal_processing_models(X_train, y_train, X_test, train_lightgbm_model)
# Parallel LightGBM
parallel_lgb_result = parallel_processing_models(X_train, y_train, X_test, train_lightgbm_model, n_jobs=n_jobs)

lgb_results = {
    "Normal LightGBM": {
        "Time (s)": normal_lgb_result['time'],
    },
    "Parallel LightGBM (Joblib)": {
        "Time (s)": parallel_lgb_result['time'],

    }
}
lgb_results_df = pd.DataFrame(lgb_results).T
print("\nLightGBM Performance Comparison:")
print(lgb_results_df)

# --- Compare XGBoost ---
print("\n--- Comparing XGBoost Processing ---")
# Normal XGBoost
normal_xgb_result = normal_processing_models(X_train, y_train, X_test, train_xgboost_model)
# Parallel XGBoost
parallel_xgb_result = parallel_processing_models(X_train, y_train, X_test, train_xgboost_model, n_jobs=n_jobs)

xgb_results = {
    "Normal XGBoost": {
        "Time (s)": normal_xgb_result['time'],

    },
    "Parallel XGBoost (Joblib)": {
        "Time (s)": parallel_xgb_result['time'],

    }
}
xgb_results_df = pd.DataFrame(xgb_results).T
print("\nXGBoost Performance Comparison:")
print(xgb_results_df)
```

```python

# Combine the training times for plotting
model_times = {
    'Normal LightGBM': normal_lgb_result['time'],
    'Parallel LightGBM': parallel_lgb_result['time'],
    'Normal XGBoost': normal_xgb_result['time'],
    'Parallel XGBoost': parallel_xgb_result['time']
}

models = list(model_times.keys())
training_times = list(model_times.values())

plt.figure(figsize=(10, 6))
bars = plt.bar(models, training_times, color=['skyblue', 'dodgerblue', 'lightcoral', 'indianred'])
plt.ylabel('Training Time (seconds)')
plt.title('Training Time Comparison of Models')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Add time values on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.01, round(yval, 2), ha='center', va='bottom')

plt.show()
```
